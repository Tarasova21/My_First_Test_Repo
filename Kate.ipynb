```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
df=pd.read_excel('TES.xlsx')
df = df.dropna()
df
```

1)Cортировка данных

```python
sorted_df = df.sort_values(by='Владелец', ascending=True) 
sorted_df
```

2)Добавление столбца к таблице

```python
sred = df['Мощность'].mean()
df1 = df.assign(Отличие_от_средней_мощности=df['Мощность'] - sred)
df1
```

3)Вывод таблицы по условию



```python
condition = df['Мощность'] <4
filtered_df = df[condition]
filtered_df
```

4)Вывод фрагмента таблицы по его расположению(строка,столбец)

```python
filtered_df = df.iloc[[2,3,6], [2,3]] 
filtered_df
```

5)Рассчёты по данным таблицы

```python
total_sales_by_category = df.groupby('Регион')['Мощность'].sum()
print(total_sales_by_category)
```

6)Построение гистограммы

```python
plt.bar(df['Регион'], df['Мощность'])
plt.gca().yaxis.set_minor_locator(plt.MultipleLocator(0.5))
plt.xticks(rotation=90)
plt.xlabel('Регион')
plt.ylabel('Мощность')
plt.title('Гистограмма мощности ТЭС в России')
```

7)Построение круговой диаграммы

```python
plt.pie(df['Мощность'], labels=df['ТЭС'])
plt.title('Круговая диаграмма мощности ТЭС')
plt.show()
```

8)Посторение графика

```python
fig = px.bar(df, x='ТЭС', y='Мощность')
fig.show()
```
