```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Загружаем данные из Excel файла
df = pd.read_excel('Угольные_разрезы.xlsx')

# Оставляем только последние 5 суток
df = df.tail(24*5*2)

# Считаем суммарное потребление обоих разрезов
df['Total'] = df['Разрез 1'] + df['Разрез 2']

# Строим график суммарного потребления
plt.plot(df['Дата/Время'], df['Total'])

# Находим локальные максимумы
max_points = []
for i in range(1, len(df['Total'])-1):
  if df['Total'][i] > df['Total'][i-1] and df['Total'][i] > df['Total'][i+1]:
    print(f"Локальный максимум: {df['Total'][i]}")
    max_points.append(df['Дата/Время'][i])

# Отмечаем локальные максимумы на графике
max_values = df.loc[df['Дата/Время'].isin(max_points)]
plt.plot(max_values['Дата/Время'], max_values['Total'], 'ro')

# Настраиваем отображение графика
plt.xlabel('Дата/Время')
plt.ylabel('Суммарное потребление, МВт')
plt.title('Суммарное потребление двух угольных разрезов')
plt.xticks(rotation=45)

# Отображаем график
plt.show()
```
